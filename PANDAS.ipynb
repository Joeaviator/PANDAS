```python
import numpy as np
import pandas as pd
```

```python
#Create the Data
weathere_data={
    "Days":["Monday", "Tuesday",None, "Wednesday", "Thursday", "Friday", "Saturday",None, "Sunday"],
    "Temp":[23,29,35,None, 29,20,13,32,None],
    "Description":["Pretty Cold",None, "Windy", "Hot Summer Day", "Chilly and Windy",None, "Very Cold", "Mild Snow", "Hot temperature"]
}
print(weathere_data)
```

```python
#Convert into a series
weathere_series=pd.Series(weathere_data)
weathere_series
```

```python
#Create a dataframe
df=pd.DataFrame(weathere_data)
df
```

```python
#Sort the data by temperature
df_sorted=df.sort_values(by='Temp')
df_sorted
```

#Observations


*   The hottest day is Wednesday with an temperature of 35℃ with a description  of "Hot Summer Day"
*   The coldest day was Saturday with a temperature of 13℃ with a description of Mild Snow
